```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path=r"C:\Users\rocks\Documents\file handling\winequality_red.csv"
wine_df=pd.read_csv(path)
wine_df
```

```python
unique_labels= wine_df['chlorides'].unique()
count=[]
for i in unique_labels:
    con=wine_df['chlorides']==i # True and False
    count.append(len(wine_df[con]))

    chlorides_df=pd.DataFrame(zip(unique_labels,count),
                              columns=['chlorides','Count'])
chlorides_df.to_csv('chlorides_df.csv',index=False)
```

```python
chlorides_df
```

```python
chlorides_vc=wine_df['chlorides'].value_counts() # series
l1=chlorides_vc.keys()
l2=chlorides_vc.values
chlorides_vc_df=pd.DataFrame(zip(l1,l2),
                             columns=['chlorides','count'])
chlorides_vc_df
```

```python
chlorides_df
```

```python
chlorides_vc_df
```

```python
plt.figure(figsize=(10,6)) # to incease the plot size
plt.bar('chlorides',
        'count',
        data=chlorides_vc_df)
plt.xlabel("chlorides") # x-axis name
plt.ylabel('count') # y-axis name
plt.title("Bar chart") # title of the chart
plt.savefig('chlorides_bar.jpg')
plt.show()
```
