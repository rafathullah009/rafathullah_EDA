create dataframes using list

```python
name=['Ramesh','suresh','sathish']
age=[30,35,40]
name,age
```

```python
pd.DataFrame()
```

```python
pd.DataFrame(zip(name,age))
```

```python
data=zip(name,age)
cols=['Name','age']
pd.DataFrame(data,columns=cols)
```

$ provide $ $index$

```python
data=zip(name,age)
cols=['Name','age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

** step-5**

$ add new column$

```python
city_names=['hyderabad','banglore','chennai']
df['city']=city_names
df
```

** steps-6**

```python
df['Name']=['swamy anna','abu','asif']
df
```

** step 7**

```python
df.drop('city',axis=1,inplace=True)
```
