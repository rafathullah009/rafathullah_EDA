### STEP-5

## B 

```python
name=['Rakesh','mallesh','kamesh']
age=[35,45,50]

df=pd.DataFrame(zip(name,age),
                columns=['Name','Age'],
                index=['A','B','C'])

city_names=['Hyderabad','Banglore','chennai']
df['city']=city_names

df.drop('city',
        axis=1,
        inplace=True)

df
```

## step-8

```python
df.drop('C', axis=0,inplace=True)

df
```
